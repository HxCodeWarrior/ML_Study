商品评论情感分析——步骤分析

```python
# 1.获取数据
import pandas as pd
data = pd.read_csv("../../Data/商品情感分析/书籍评价.csv", encoding="gbk")
```

```python
data
```

2.数据基本处理

```python
# 2.1.取出内容列，对数据进行分析
data_content = data["内容"]
data_content
```

```python
# 2.2.判定评判标准  1---good   0---bad
data.loc[data.loc[:, '评价'] == "好评", "评论编号"] = 1
data.loc[data.loc[:, '评价'] == "差评", "评论编号"] = 0
```

```python
good_or_bad = data["评价"].values
```

```python
# 2.3.选择停用词
stopwords = []
with open("../../Data/商品情感分析/stopwords.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
    print(lines)
    for temp in lines:
        line = temp.strip()
        # print(line)
        stopwords.append(line)

# 对停用词进行去重
stopwords = list(set(stopwords))
```

```python
# 2.4.把内容处理，转化成标准格式
import jieba

comment_list = []
for temp in data_content:
    print(temp)
    #  对文本进行切割
    # cut_all 参数默认为False，所使用cut方法默认为精简模式
    seg_list = jieba.cut(temp, cut_all=False)
    print(seg_list)
    seg_str = ",".join(seg_list)    # 字符串拼接
    print(seg_str)
    comment_list.append(seg_str)    # 目的是转换成列表形式
```

```python
# 2.5.统计词的个数
# 进行统计词个数
# 实例化对象
# CountVectorizer类会将文本中词语语转换为词频矩阵
from sklearn.feature_extraction.text import CountVectorizer
con = CountVectorizer(stop_words = stopwords)
# 进行词数统计
X = con.fit_transform(comment_list)     # 通过fit_transform函数计算各个词语出现的次数
name = con.get_feature_names_out()     # 通过get_feature_names_out可获取磁带中所有文本的关键字
print(X.toarray())
print(name)
```

```python
# 2.6.准备训练集和测试集
# 准备训练集  这里将文本前10行当作训练集，后三行当作测试集
x_train = X.toarray()[:10, :]
y_train = good_or_bad[:10]
# 准备测试集
x_test = X.toarray()[10:, :]
y_test = good_or_bad[10:]
```

3.模型训练

```python
# 3.模型训练
from sklearn.naive_bayes import  MultinomialNB
mb = MultinomialNB(alpha=1)
# 训练数据
mb.fit(x_train, y_train)
# 预测数据
y_predict = mb.predict(x_test)
print("预测值：", y_predict)
print("真实值：", y_test)
```

4.模型评估

```python
# 4.模型评估
mb.score(x_test, y_test)
```
