1 什么是机器学习

1.1 含义：机器学习是从数据中自动分析获得模型，并利用模型对未知数据进行预测

    （1）历史数据---->(训练)---->模型
    （2）新的数据---->(输入)---->模型---->(预测)---->未知属性

1.2 类比人类学习

    （1）历史经验---->(归纳总结)---->规律
    （2）新的问题---->(输入)---->规律---->(预测)---->未来

2 机器学习工作流程

    用户数据---->(输入)---->模型对数据处理---->服务

    模型对数据的处理
        （1）数据基本处理
        （2）特征工程
        （3）机器学习
        （4）模型评估
            * 结果达到要求，上线服务
            * 结果没达到要求，重新上述过程

    2.1 获取数据。包括表格数据、图像数据、声音数据等等。
        1. 数据集介绍
            （1）样本：表格中的一行数据。
            （2）特征：表格中的一列数据。
            （3）目标值：对数据进行判断的一列。目标值分为连续的目标值、离散的目标值。
        ！！！注意：有些数据没有目标值，有些数据有目标值
        2.数据类型的构成
            （1）数据类型一：特征值+目标值
            （2）数据类型二：只有特征值，没有目标值
        3.数据分割
            （1）机器学习一般的数据集合会划分为两个部分：
                    * 训练数据：用于训练、构建模型
                    * 测试数据：在模型检验时使用，用于评估模型是否有效
            （2）划分比例：
                    * 训练集：70% 80% 75%
                    * 测试集：30% 20% 25%
        4.数据处理：
            （1）数据要有代表性，否则必然会过拟合
            （2）对于分类问题，数据偏斜不能过于严重，不同类别的数据数量不要有数量级的差距
            （3）要对数据的量级有一个评估，多少个样本，多少个特征，可以估算出其对内存的消耗程度，判断训练过程中内存是否能够放得下。如果放不下就得考虑改进算法或者使用一些降维的技巧。如果数据量实在太大，就要考虑分布式。

    2.2 数据基本处理
        即对数据进行缺失值、去除异常值等处理
        1.特征处理
            特征处理、数据清洗是很关键的步骤，往往能够使得算法的效果和性能得到显著提高。归一化、离散化、因子化、缺失值处理、去除共线性等等，数据挖掘过程中很多时间都花费在这个上面。这些工作简单可复制、收益稳定可预测，是机器学习西的基础必备步骤。
        2.特征筛选
            筛选出显著特征、摒弃非显著特征，需要机器学习工程师反复理解业务，这时很多结果有决定性的影响。特征选择好了，非常简单的算法也能得出良好、稳定的结果，这需要运用特征有效性分析的相关技术，如相关系数、卡方检测、平均互信息、条件熵、后验概率、逻辑回归权重等方法。

    2.3 特征工程（Feature Engineering）
        1.特征工程介绍
            （1）使用也能够专业背景知识和技巧处理数据，使得特征在机器学习算法上发挥更好的作用的过程
            （2）意义：会直接影响机器学习的效果
            （3）注意：业界广泛流传：数据和特征决定了机器学习的上线，而模型和算法只是逼近这个上限而已
        2.特征工程包含内容
            （1）特征提取：将任意数据（如：文本或者图像）转换为可用于机器学习的数字特征
            （2）特征预处理：通过一些转换函数将特征数据转换成更加适合算法模型的特征数据过程。简单来说就是将一些数据范围不同的特征数据，转化为数据范围相同的数据，但是每组特征值表示的含义不同。
            （3）特征降维：指在某些限定条件下，降低随机变量（特征）个数，得到一组“不相关”主变量的过程。例如：将3D地球模型，转化为2D全球地图。

    2.4 机器学习  选择合适的算法对模型进行训练（详解见）
        训练模型与调优，知道这一步才用到我们上面说的算法进行训练，现在很多算法都能够封装成黑盒供人们使用。但是真正考验水平的是调整这些算法的（超）参数，使得结果变得更加优良。这需要我们对算法的原理有更深入的理解。理解越深入，就越能发现问题的症结，提出良好的调优方案。

    2.5.模型评估  队训练好的模型及性能评估（详解见）
        （1）模型诊断。
        如何确定模型体哦啊有的方向与思路，这就需要对模型进行诊断的技术.
        过拟合、欠拟合判断是模型诊断中至关重要的一步。常见的方法：交叉验证、绘制学习曲线等。
        ！过拟合的基本调优思路时增加数据量，降低模型复杂度。
        ！欠拟合的基本调优思路是提高特征数量和质量，增加模型复杂度。
        （2）误差分析
        误差分析也是机器学习至关重要的步骤。通过观察误差样本全面分析产生误差的原因：是参数的问题还是算法选择的问题，是特征的问题还是数据本身的问题……
        
        诊断后的模型需要进行调优，调优后的模型需要重新进行诊断，这是一个反复迭代不断逼近的过程，需要不断地尝试，进而达到最优状态

    2.6.模型融合
        * 一般来说，模型融合后都能使效果有一定提，而且效果很好。
        工程上，主要提升算法准确度的方法是分别在模型的前端（特征清洗，不同的采样模式）
        与后端（模型融合）上下功夫。因为他们比较标准可复制，效果比较稳定，而且直接调参的工作不会很多，毕竟大量数据训练起来太慢了，而且效果难以保证    

    2.7.上线运行
        这一部分的主要内容跟工程实现的相关性比较大。功臣杀害那个是结果导向，模型在线上运行的效果直接决定模型的成效。
        不单纯包括其准确程度、误差等情况，还包括其运行的速度（时间复杂度）、资源消耗程度（空间复杂度）、稳定性是否可接受。

这些工作流程主要是工程实践上总结出的一些经验，并不是每个项目都包含完整的一个流程。
这里的部分只是一个指导性的说明，只有自己的实践，多积累项目经验，才有自己更深入的认识
