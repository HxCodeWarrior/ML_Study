1.流程分析

    （1）导入工具库
    （2）创建数据集并显示
    （3）应用KMeans

```python
# 0.导包
import matplotlib.pyplot as plt
from sklearn.datasets._samples_generator import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score
```

```python
# 1.创建数据集
# x--样本特征，y--样本簇类别，共1000个样本，每个样本4个特征，共4个簇
# 簇中心在[-1, 1],[0, 0],[1, 1],[2, 2]
# 簇方差分别为：[0.4, 0.2, 0.2, 0.2]
x, y = make_blobs(
    n_samples= 1000,    # 生成样本数
    n_features= 2,      # 样本维度
    centers = [[-1, 1],[0, 0],[1, 1],[2, 2]],   # 生成的样本簇中心
    cluster_std = [0.4, 0.2, 0.2, 0.2],     # 生成样本方差
    shuffle = True,
    random_state = 9)
```

```python
# 2.数据可视化
plt.scatter(x[:,0], x[:, 1], marker="o")
plt.show()
```

```python
# 3.使用K-Means进行聚类，并使用CH方法评估
y_pred = KMeans(n_clusters=2, random_state=9).fit_predict(x)
# 分别尝试n_cluses=2/3/4查看聚类效果
plt.scatter(x[:, 0], x[:, 1], c=y_pred) 
#x[:, 0]前一个取所有，后一个取第一个维度
#x[:, 0]前一个取所有，后一个取第二个维度
plt.show()
```

```python
# 用Calinski-Harabasz Index评估聚类分数
calinski_harabasz_score(x, y_pred)
```

```python
# 3.使用K-Means进行聚类，并使用CH方法评估
y_pred = KMeans(n_clusters=2, random_state=9).fit_predict(x)
# 分别尝试n_cluses=2/3/4查看聚类效果
plt.scatter(x[:, 0], x[:, 1], c=y_pred) 
#x[:, 0]前一个取所有，后一个取第一个维度
#x[:, 0]前一个取所有，后一个取第二个维度
plt.show()
```

```python
# 用Calinski-Harabasz Index评估聚类分数
calinski_harabasz_score(x, y_pred)
```
