其他距离公式

    1.标准化欧式距离（Standardized EuclideanDistance）：
        （1）介绍：标准化欧式距离是针对欧式距离的缺点而作的一种改进
        （2）思路：既然各个维度分量的分布不一样，那就先将各个分量都“标准化”到均值、方差等
        （3）公式：
                Sk表示各个维度的标准差
                标准欧式距离：
                        d12=(∑( (x1k - x2k) / Sk) ^ 2)^2
                注意：如果将方差的倒数看成一个权重，也可称之为加权欧式距离（Weighted Euclidean Distance）

    2.余弦距离（Codine Distance）
        （1）介绍：几何中，夹角余弦值可用来衡量两个方向的差异，在机器学习中，借用这一概念来衡量样本向量之间的差异
        （2）公式介绍：
                二维空间中向量A(x1, y1)与向量B(X2, y2)的夹角余弦公式：
                    cos = (x1*x2 + y1*y2) / [(x1^2 + y1^2)^0.5 + (x2^2 + y2^2)^0.5]
                n维空间中样本点a(x11, x12,……)与b(x21, x22，……)的夹角余弦公式：
                    cos(θ) = (a*b) / (|a|*|b|)
        （3）解释：夹角余弦值的取值范围为[-1, 1],
                 余弦值越大表示两个向量的夹角越小，余弦值越小表示两个向量的夹角越大。
                 当两个向量的方向重合时余弦取最大值1，当两个向量的方向完全相反时余弦取最小值-1

    3.汉明距离（Hamming Distance）【了解】
        （1）介绍：两个等长字符串s1与s2的汉明距离为：将其中一个变为另外一个所需要作的最小字符替换次数
        （2）汉明重量
            （2.1）汉明重量是字符串相对于同样长度的0字符串的汉明距离，也就是说，
            它是字符串中非零的元素个数：对于二进制的字符串来说，就是1的个数。
            所以11101的汉明重量是4.因此，如果向量空间中的元素a和元素b之间的汉明距离等于他们汉明重量的差a-b
        （3）应用
            汉明重量分析在包括信息论、编码理论、密码学等领域都有应用。
            比如在信息编码的过程中，为了增强容错性，应使得编码间的最小汉明距离尽可能大。
            但是如果比较两个不同长度的字符串，不仅要进行替换，而要进行插入与删除的运算，
            在这种场合下，通常使用更加复杂的编辑距离等算法。
        （4）举例
            "1011101" 和 "1001001"之间的汉明距离：2
            "2143896" 和 "2233796"之间的汉明距离：3
            "toned" 和 "roses"之间的汉明距离：3

    4.杰卡德距离（Jaccard Distance）【了解】
        （1）杰卡德相似系数（Jaccard similarity conefficient）
            两个集合A和B的交元素在A，B的并集中所占的比例，称为两个集合杰卡德相似系数，用符号J（A，B）表示：
                J(A,B) = |A∩B| / |A∪B|
        （2）杰卡德距离（Jaccard Distance）
            与杰卡德相似系数相反，用两个集合中不同元素所占元素的比例来衡量两个集合的区分度：
                Jσ(A,B) = 1 - J(A,B)
                        = （|A∪B| - |A∩B|） / |A∪B|

    5.马氏距离（Mahalanobis Distance）【了解】
        （1）马氏距离是基于样本分布的一种距离
        （2）马氏距离由印度统计学家马哈拉诺比斯提出，表示数据的协方差距离，
            它是一种有效的计算两个位置样本集的而相似度的方法
        （3）与欧式距离不同的是，它考虑到各种特性之间的联系，即独立于测量尺度
        （4）马氏距离定义：
                设总体G为m维总体（考察m个指标）
                均值向量维μ=（μ1，μ2，……，μm，）
                协方差阵∑=（σij）
              则样本X=（X1, X2, ……，Xm,）与总体G的马氏距离定义为：
                d^2(X, G) = (X - μ)' ∑^-1 (X - μ)
              当m=1时，d^2(x, G) = (x - μ)'(x - μ) / (σ^2)
                                = (x - μ)^2 / (σ^2)
        （5）马氏距离特性：
            （a）量纲无关。排除变量之间的相关性干扰。
            （b）马氏距离的计算是建立在总体样本的基础上的。
                如果拿相同的两个样本，放入两个不同的总体中，最后计算得出的两个样本间的
                马氏距离通常是不同的，排除这两个总体的协方差矩阵碰巧相同
            （c）计算马氏距离过程中，要求总体样本数大于样本的维数，
                否则得到的总体样本协方差逆矩阵不存在，这种情况下，用欧式距离计算即可
            （d）话有一种情况，满足了条件总体样本数大于样本的维数，但是协方差矩阵的逆
                矩阵仍然不存在，
                比如三个样本点（3，4），（5，6），（7，8），
                这种情况是因为这三个样本在其所处的二维空间平面内共线，这种情况下也采用欧式距离计算。
