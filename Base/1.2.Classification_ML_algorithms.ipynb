机器学习算法分类

    1.监督学习
        （1）定义：输入数据是由输入特征和目标值所组成。
        （2）函数的输出：是一个连续的值——回归
                  是有限个离散值——分类
        1.1 回归问题
            例如：预测房价、预测天气，根据样本集拟合出一条连续曲线
        1.2 分类问题
            例如：根据肿瘤特征判断良性还是恶性，得到的结果是“良性”或者“恶性”，是离散的
        1.3 监督学习训练方式
            大量未被标记数据（没有目标值的数据）----->获得未被标记的数据---->专家预测---->得到带标记的数据---->预测训练模型---->监督学习模型

    2.无监督学习
        2.1 无监督学习简介
        （1）定义：输入数据由输入特征组成，没有目标值
        （2）输入数据没有被标记，也灭有确定的结果，样本数据类别未知
        （3）需要根据样本间的相似性对样本集进行类别划分
        
        2.2 有监督、无监督算法对比
            （1）监督学习——输入的数据有特征值，有目标值
                * 目标值连续：回归
                * 目标值离散：分类
            （2）无监督学习——输入的数据有特征值，无目标值

    3.半监督学习
        3.1 定义：训练集同时包含有标记样本数据和未标记样本数据（即同时包含有目标值数据和无目标值数据）
        3.2 半监督学习训练方式
            大量的数据---->得到少部分未被标记的数据---->专家预测---->少量带标记数据----（+大量的未被标记的数据）---->训练初步模型

    4.强化学习
        4.1 定义：实质是make decision的问题，即自动进行决策，并且可以做连续决策
        
        ---举例：小孩想要走路，但是在这之前，他需要先站起来，站卡里之后还要保持平衡，接下来还要先迈出一条腿，是左腿还是右腿，迈出一步之后还要迈出下一步。
        小孩：agent，他试图通过采取行动（即走路）来操纵环境（行走的表面），并且从一个状态转变到另一个状态（即他走的每一步），当他完成任务的子任务（即走了几步）时，孩子得到奖励（给巧克力），并且当他不能走路时，就不会给巧克力。
        
        4.2 主要包含五个元素：
            （1）agent
            （2）action —— 行动
            （3）reward —— 奖励
            （4）environment —— 环境
            （5）observation —— 
        4.4 强化学习的目标：获得最多的累计奖励
        4.5 监督学习和强化学习对比
            （1）监督学习
                （1.1）反馈映射：输出的是之间的关系，可以告诉算法什么样的输入对应着什么样的输出
                （1.2）反馈时间：做了比较坏的选择会立即反馈个算法
                （1.3）输入特征：输入是独立同分布的
            （2）强化学习
                （2.1）反馈映射：输出的是给机器的反馈（reward function），及用来判断这个行为是好还是坏
                （2.2）反馈时间：结果反馈有延时，有时候可能需要走了很多步以后才知道以前的某一步的选择是好还是坏
                （2.3）输入特征：面对的输入总是在变化，每当算法做出一个行为，它影响下一次决策的输入          

    5.小结
    objects              INPUT             OUTPUT           GOAL          
    监督学习               有标签             有反馈           预测结果
    无监督学习             无标签              无反馈         发现潜在结构
    半监督学习        部分有标签、部分无标签      有反馈       降低数据标记的难度
    强化学习           决策流程及激励系统      一系列行动        长期利益最大化
