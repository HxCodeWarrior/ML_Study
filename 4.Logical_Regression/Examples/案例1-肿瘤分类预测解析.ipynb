```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
```

```python
breast_cancer_wisconsin_original = pd.read_csv("../../Data/Breast_Cancer_Wisconsin/breast-cancer-wisconsin.data")
```

```python
breast_cancer_wisconsin_original
```

    可以看到，该数据集并没有在第一行展示索引,需要我们手动添加索引

```python
columns_names = ["Sample_code_number", "Clump_thickness", "Uniformity_of_cell_size",
         "Uniformity_of_cell_shape", "Marginal_adhesion", "Single_epithelial_cell_size",
         "Bare_nuclei", "Bland_chromatin", "Normal_nucleoli", "Mitoses", "Class"
         ]
breast_cancer_wisconsin = pd.read_csv(
    "../../Data/Breast_Cancer_Wisconsin/breast-cancer-wisconsin.data",
    names=columns_names
)
```

```python
breast_cancer_wisconsin
```

    可以看到，在目标值列，若class=2则为良性，若clas=4则为恶性

```python
# 2.数据基本处理-处理缺失值
# 由于缺失值很少，样本数据集很大，我们就直接删掉
data = breast_cancer_wisconsin.replace(to_replace="?", value=np.nan)
data = data.dropna()
```

```python
data
```

```python
# 2.数据基本处理-确定目标值
x = data.iloc[:, 1:10]
y = data["Class"]
```

```python
# 特征值
x
```

```python
# 目标值
y
```

```python
# 2.数据基本处理—数据分割
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
# 3.特征工程-标准化
transfer = StandardScaler()
x_train = transfer.fit_transform(x_train)
x_test = transfer.fit_transform(x_test)
```

```python
# 4.机器学习--逻辑回归
estimator = LogisticRegression()
estimator.fit(x_train, y_train)
```

```python
# 5.模型评估
y_predict = estimator.predict(x_test)
ret = mean_squared_error(y_test, y_predict)
score = estimator.score(x_test, y_test)
print("模型预测值：\n", y_predict)
print("模型均值方差：",ret)
print("模型准确率：",score)
```
